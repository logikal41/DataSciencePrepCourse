```python
# import all necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1 - Normal Distribution

```python
normal = np.random.normal(10,1.5,100)
```

```python
# Print out a histogram
plt.hist(normal)
plt.show()
```

```python
# Add vertical lines for the mean and standard deviations on the hisogram plot
normal_mean = normal.mean()
normal_stdev = normal.std()
plt.axvline(x=normal_mean, color="g", linestyle="solid")
plt.axvline(x=(normal_mean - normal_stdev), color="g", linestyle="dashed")
plt.axvline(x=(normal_mean + normal_stdev), color="g", linestyle="dashed")
plt.hist(normal)
plt.show()
```

  The descriptive statistics defintely provide useful information about the normal distribution. The datapoints cluster at the mean and ~68% of all data points fall between 1 standard devation from the mean.

The descriptive statistics defintely provide useful information about the normal distribution. The datapoints cluster at the mean and ~68% of all data points fall between 1 standard devation from the mean. Adding two normal distributions together appears to create a third normally distributed plot.
